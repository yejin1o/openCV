```python
Img1 = cv2.imread('/content/gdrive/My Drive/ColabIP/milkdrop.bmp')

RGVB_Img = cv2.cvtColor(Img1, cv2.COLOR_BGR2RGB)
```

```python
gImg = cv2.cvtColor(RGVB_Img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
cv2_imshow(gImg)
```

```python
kernel = np.ones((3,3), np.uint8)
dilateImg = cv2.dilate(gImg, kernel, iterations = 1)
cv2_imshow(dilateImg)
```

```python
gBin = np.zeros((gImg.shape[0], gImg.shape[1]), dtype=np.ubyte)

threshold = 130

for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    if(gImg[h,w] > threshold):
      gBin[h,w] = 255

cv2_imshow(gBin)
```

```python
kernel = np.ones((3,3), np.uint8)
erodeImg = cv2.erode(gBin, kernel, iterations = 3)
cv2_imshow(erodeImg)
```
