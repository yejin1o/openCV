```python
img1 = cv2.imread('/content/gdrive/My Drive/ColabIP/lena256.jpg')
#img1 = cv2.imread('/content/gdrive/My Drive/ColabIP/fce(salt_pepper noise).bmp')
RGB_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
cv2_imshow(gImg)
#plt.imshow(gImg1, cmap='gray')
```

```python
outImg = cv2.GaussianBlur(gImg, (3,3), 3)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (7, 7), 3)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (15, 15), 3)
cv2_imshow(outImg)
```

```python
mean_Img = cv2.blur(gImg, (5,5))
cv2_imshow(mean_Img)
```

```python
edge_Img = cv2.addWeighted(gImg, 1.0, mean_Img, -1.0, 0)
cv2_imshow(edge_Img)
```

```python
out_Img = cv2.addWeighted(gImg, 1.0, edge_Img, 1.0, 0)
cv2_imshow(out_Img)
```
