```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/maru.JPG')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg1 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg1.shape)
cv2_imshow(gImg1)
#plt.imshow(gImg1, cmap='gray')
```

```python
gOutImg = np.zeros((gImg1.shape[0], gImg1.shape[1]))
```

```python
for h in range(0, gImg1.shape[0]):
  for w in range(0, gImg1.shape[1]):
    val = gImg1[h,w] % 64
    val = val * 4
    gOutImg[h,w] = val

#plt.imshow(gOutImg.cmap = 'gray')
cv2_imshow(gOutImg)
```

```python
equalVal = 128
width = 5

min = equalVal - width
max = equalVal + width
```

```python
for h in range(0, gImg1.shape[0]):
  for w in range(0, gImg1.shape[1]):
    if(gImg1[h,w] > min and gImg1[h,2] < max):
      gOutImg[h,w] = 0
    else:
      gOutImg[h,w] = gImg1[h, w]

plt.imshow(gOutImg, cmap = 'gray')
```

```python
for h in range(0, gImg1.shape[0]):
  for w in range(0, gImg1.shape[1]):
    val = 255*(gImg1[h,w]/128 -1)*(gImg1[h,w]/128 -1)
    gOutImg[h,w] = val

cv2_imshow(gOutImg)
```

```python
planeNumber = 1         # 7 or 6  or 3 번째로 수정 후 출력값 비교
planeValue = 1

for i in range(planeNumber-1):
  planeValue = planeValue * 2
```

```python
for h in range(0, gImg1.shape[0]):
  for w in range(0, gImg1.shape[1]):
    imVal = (int)(gImg1[h,w] / 128)          # 8번째 비트가 1
    if(imVal % 2 == 0):
      gOutImg[h,w] = 0
    else:
      gOutImg[h,w] = 255

cv2_imshow(gOutImg)
```
