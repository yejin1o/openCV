```python
def threshold_update(h, w, img, output, T1):
  sum1 = 0
  count1 = 1
  sum2 = 0
  count2 = 1

  for h in range(img.shape[0]):
    for w in range(img.shape[1]):
      if(output[h,w] == 255):
        sum1 = sum1 + img[h,w]
        count1 = count1 + 1
      else:
        sum2 = sum2 + img[h,w]
        count2 = count2 + 1

avg1 = sum1 / count1
avg2 = sum2 / count2

T2 = np.int32(avg1 + avg2)/2
return T2
```
